```python
def generate_remaining_IDs(all_indices,removable_indeces):
    """
    usually :  all_indices = range(0,...)
    """
    new_indeces = []
    for i in all_indices:
        if i not in removable_indeces:
            new_indeces.append(i)

    return new_indeces
```

```python
def reduce_adjacency_mat(old_mat,new_indices):#all_indices,remove_indices):
    # new_indices = generate_remaining_IDs(all_indices,remove_indices)
    print('initial dimention :',np.shape(old_mat))
    new_mat = np.zeros((len(new_indices),len(new_indices)))

    for x,i in enumerate(new_indices):
        for y,j in enumerate(new_indices):
            new_mat[x,y] = old_mat[i,j]

    print('final dimention :',np.shape(new_mat))
    return new_mat
```

```python
def reduce_feature_mat(old_feature_mat,new_indices):#all_indices,remove_indices):
    print('initial dimention :',np.shape(old_feature_mat))
    # new_indices = generate_remaining_IDs(all_indices,remove_indices)
    mat = old_feature_mat.tolist().copy()

    modified_list = [mat[i] for i in new_indices]


    print('final dimention :',np.shape(modified_list))
    return np.array(modified_list)
```

```python
import pandas as pd

def reduce_df(df,remove_indices):
    print('initial dimention :',df.shape[0])
    # Rows to remove (indices)
    rows_to_remove = [df.index[i] for i in remove_indices]

    # Remove the specified rows
    modified_df = df.drop(rows_to_remove)

    print('initial dimention :',modified_df.shape[0])
    return modified_df
```
